## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Establish bins and group names for ages
bins = [0,9,14,19,24,29,34,39,100]
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#Categorize the existing players using the age bins
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
grouped_data_byAge=purchase_data.groupby(["SN"])
age_group_valuects =grouped_data_byAge["Age Group"].value_counts()

# put value counts into data frame
age_group_valuects = pd.DataFrame(age_group_valuects)

age_group_valuects = age_group_valuects.rename(columns={"Age Group":"Counts"})
age_group_valuects = age_group_valuects.reset_index(level=["Age Group"])
age_group_valuects = pd.DataFrame(age_group_valuects["Age Group"].value_counts())
age_group_valuects = age_group_valuects.rename(columns={"Age Group":"Total Count"})
age_group_valuects = age_group_valuects.reindex(["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"])

#adding a column to the dataframe age_group_counts_df called "Percentage of Players"
# and will figure out the formula for you.

sum_age_valuects = age_group_valuects["Total Count"].sum()

age_group_valuects["Percentage of Players"] = (age_group_valuects["Total Count"]/sum_age_valuects)

#Formatting the value in the summary table
age_group_valuects['Percentage of Players'] = pd.to_numeric(age_group_valuects['Percentage of Players'])
age_group_valuects['Percentage of Players'] = age_group_valuects['Percentage of Players'].map("{:.2%}".format)

age_group_valuects
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# purchase_data data frame bin by age
purchase_data
#Calculating Purchase Count by Age Group.The Series is named as "pur_cnt"
ag_data_byb=pd.Series(purchase_data["Age Group"].value_counts(),name="pur_cnt")

#calculating the average purcahse by Age Group.The Series is named as "pur_avg"
ag_data_avg=pd.Series(round(purchase_data.groupby("Age Group")["Price"].mean(),2),name="pur_avg")

#calculating the total purcahse by Age Group.The Series is named as "tot_pur"
ag_data_sum=pd.Series(round(purchase_data.groupby("Age Group")["Price"].sum(),2),name="tot_pur")

age_group=purchase_data.groupby("Age Group")
#ag_data_byg,ag_data_avg,ag_data_sum

purch_perSN = pd.DataFrame(age_group["SN"].unique())
total_purch_perSN = purch_perSN["SN"].str.len()

ag_avg_ppp = round((ag_data_sum / total_purch_perSN), 2)

#ag_data_byb,ag_data_avg,ag_data_sum,ag_avg_ppp

#Creating an output dataframe for Purchasing Analysis by Age using Concat function on axis 1
purch_age_summdf = pd.concat([ag_data_byb,ag_data_avg,ag_data_sum,ag_avg_ppp], axis=1, join='outer')

#Rename the columns
purch_age_summdf=purch_age_summdf.rename(columns={"pur_cnt":"Purchase Count","pur_avg":"Average Purchase Price","tot_pur":"Total Purchase Value",0:"Average Total Purchase per Person"})


#formatting the columms
purch_age_summdf["Average Purchase Price"] = purch_age_summdf["Average Purchase Price"].map("${:.2f}".format)
purch_age_summdf["Total Purchase Value"] = purch_age_summdf["Total Purchase Value"].map("${:.2f}".format)
purch_age_summdf["Average Total Purchase per Person"] = purch_age_summdf["Average Total Purchase per Person"].map("${:.2f}".format)

#print the dataframe
purch_age_summdf
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# purchase_data.columns
# item_cnt=purchase_data['Item Name'].value_counts()
# purchase_data.loc[purchase_data['Item Name']=="Oathbreaker, Last Hope of the Breaking Storm"]
item_df=purchase_data[["Item ID", "Item Name", "Price"]].copy()

gp_item_df=item_df.groupby(["Item ID", "Item Name"])
item_ct=gp_item_df.count()

# calculate count of poplular items purchased and store into data frame
item_purch_ct_df = pd.DataFrame(item_ct)

# rename the column
item_purch_ct_df = item_purch_ct_df.rename(columns={"Price": "Purchase Count"})

#item_purch_ct_df.tail()
item_value=gp_item_df.sum()

# calculate purchase value and store into data frame
item_value_df = pd.DataFrame(item_value)

# rename the column
item_purch_val_df = item_value_df.rename(columns={"Price": "Total Purchase Value"})
item_purch_val_df.head()

# calculate item price and store into a dataframe

item_price = (item_value / item_ct)
#type(item_price_df)
item_price_df = pd.DataFrame(item_price)

# rename column
item_price_df = item_price_df.rename(columns={"Price":"Item Price"})
#item_price_df.head()

#concat all three dataframes  by the column axis (axis=1)
item_summary_table_df=pd.concat([item_purch_ct_df,item_price_df,item_purch_val_df],axis=1)
item_summary_table_df.head()

# sort the data by descending
item_summary_table_df = item_summary_table_df.sort_values(by='Purchase Count', ascending=False)

#format the dataframe columns
item_summary_table_df["Item Price"] = item_summary_table_df["Item Price"].map("${:.2f}".format)
item_summary_table_df["Total Purchase Value"] = item_summary_table_df["Total Purchase Value"].map("${:.2f}".format)

item_summary_table_df.head()
```
